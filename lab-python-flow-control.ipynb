```python
#1. Look at your code from the lab data structures, and improve repeated code with loops.

products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
for product in products:
    quantity = int(input(f"Enter quantity for {product}: "))
    print(f"The quantity of {product}: {quantity} ")
    inventory[product] = quantity

print("Inventory=", inventory)
```

```python
# Prompt the user to enter the name of a product that a customer wants to order
# Add the product name to the "customer_orders" set.
# Ask the user if they want to add another product (yes/no).
# Continue the loop until the user does not want to add another product.

customer_orders = set()

for product in products:
    product = input(f"Enter the name of a product you want to order from this list: {products}: ")
    customer_orders.add(product)
    user_input = input("Do you want to continue? (yes/no): ")
    if user_input == "no":
        break

print(customer_orders)
```

```python
#3. Instead of updating the inventory by subtracting 1 from the quantity of each product, 
# only do it for  the products that were ordered (those in "customer_orders").

updated_inventory = {}

for item in customer_orders:
    if item in inventory:
        updated_inventory[item] = inventory[item] -1
        
print(updated_inventory)


```
